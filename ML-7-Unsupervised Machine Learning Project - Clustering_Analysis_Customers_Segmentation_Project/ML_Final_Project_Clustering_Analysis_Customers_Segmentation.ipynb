# Cluster Analysis

Cluster analizde iki farklı yöntem deneyeceğiz ve bunları karşılaştıracağız.

Birincisi **K-Means algoritması** diğeri ise **hiyerarşik kümeleme**. 

Önce KMeans ile başlıyoruz.

**KMeans Clustering Teorisi**

- Amaç gözlemleri birbirlerine olan benzerliklerine göre kümelerine ayırmaktır.

- Kaç küme oluşturacağını baştan söyleyeceğiz. 

- Öyle bir k değeri belirlemeliyim ki veri setimin içerisindeki kümelenme değişimini minumum hata ile bana ortaya koymuş olsun.

- Minumum hata; kendi içindeki küme içi hataların hesaplanmasından sonra bütün kümelerin içindeki hataların toplanmasıdır.

- Her küme için rastgele centroid atar. Sonrasinda iteratif olarak güncelleme yaparak bu centroidler kümelerin tam ortasına -atanır.

- Bu işlemi neye göre yapar; bütün gözlem noktalarının bütün centroidlere olan uzaklığını hesaplar. En yakınları bir küme yapar. 

- Bir küme içerisinde verilerimizin centroide olan uzaklıklarına göre kümemiz geniş veya dar olabilir.

**Aşamaları**; 

- Kaç küme oluşturacağımızı belirtiriz.
- Küme sayısına göre random olarak ilk centroidler atanır.
- Data gözlemlerinin centroidlere olan uzaklıkları hesaplanarak, bu gözlemleri en yakın oldukları centroidlere ait kümelere atar.
- Atadığı gözlemlerle centroidler arasındaki mesafeye bakıp centroidlerin konumunu günceller.
- İterasyona devam edip grupları oluşturur.

Farklı değişkenler kullanarak birçok küme analizi yapabiliriz. Her bir küme analizi için en fazla iki değişken kullanırsak, küme etiketlerini daha net bir şekilde tanımlayabiliriz.

İlk olarak, K-Means algoritması bizden küme sayısını (n_clusters) belirlemenizi bekler. Her küme analizi için en uygun küme sayısını çeşitli şekillerde belirleyebiliriz. Bu durumda aşağıda da bahsedeceğimiz Elbow Metodunu kullanacağız.

Son olarak, her analizde farklı bilgiler elde edilir. Bu nedenle her küme analizi sonucunda farklı labellama yapılmalıdır.

- İki hipotez ortaya konulur. Null hipotezi datalarımızın uniform yani non-random olduğunu, alternatif hipotezimiz ise datalarımızın random olarak dağıldığını söyler.

```python
# normalde bu fonksiyonumuz  pyclustertend'ın hopkins kütüphanesinde var.Ancak oluşabilecek hatalara karşı 
# doğrudan fonksiyonun kendisini kullanacağız . Bundan dolayı pyclustertend'ı install etmemize gerek yoktur! 
                                                #!pip install pyclustertend --user
                                                # from pyclustertend import hopkins

from sklearn.neighbors import BallTree
import numpy as np
import pandas as pd
def hopkins(data_frame, sampling_size):
    """Assess the clusterability of a dataset. A score between 0 and 1, a score around 0.5 express
    no clusterability and a score tending to 0 express a high cluster tendency.
    Parameters
    ----------
    data_frame : numpy array
        The input dataset
    sampling_size : int
        The sampling size which is used to evaluate the number of DataFrame.
    Returns
    ---------------------
    score : float
        The hopkins score of the dataset (between 0 and 1)
    Examples
    --------
    >>> from sklearn import datasets
    >>> from pyclustertend import hopkins
    >>> X = datasets.load_iris().data
    >>> hopkins(X,150)
    0.16
    """
    if type(data_frame) == np.ndarray:
        data_frame = pd.DataFrame(data_frame)
    # Sample n observations from D : P
    if sampling_size > data_frame.shape[0]:
        raise Exception(
            'The number of sample of sample is bigger than the shape of D')
    data_frame_sample = data_frame.sample(n=sampling_size)
    # Get the distance to their neirest neighbors in D : X
    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(data_frame_sample, k=2)
    data_frame_sample_distances_to_nearest_neighbours = dist[:, 1]
    # Randomly simulate n points with the same variation as in D : Q.
    max_data_frame = data_frame.max()
    min_data_frame = data_frame.min()
    uniformly_selected_values_0 = np.random.uniform(min_data_frame[0], max_data_frame[0], sampling_size)
    uniformly_selected_values_1 = np.random.uniform(min_data_frame[1], max_data_frame[1], sampling_size)
    uniformly_selected_observations = np.column_stack((uniformly_selected_values_0, uniformly_selected_values_1))
    if len(max_data_frame) >= 2:
        for i in range(2, len(max_data_frame)):
            uniformly_selected_values_i = np.random.uniform(min_data_frame[i], max_data_frame[i], sampling_size)
            to_stack = (uniformly_selected_observations, uniformly_selected_values_i)
            uniformly_selected_observations = np.column_stack(to_stack)
    uniformly_selected_observations_df = pd.DataFrame(uniformly_selected_observations)
    # Get the distance to their neirest neighbors in D : Y
    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(uniformly_selected_observations_df, k=1)
    uniformly_df_distances_to_nearest_neighbours = dist
    # return the hopkins score
    x = sum(data_frame_sample_distances_to_nearest_neighbours)
    y = sum(uniformly_df_distances_to_nearest_neighbours)
    if x + y == 0:
        raise Exception('The denominator of the hopkins statistics is null')
    return x / (x + y)[0]
```

**Elbow Methodu**

- Optimal küme sayısını belirlemede kullanıyoruz.
- Farklı k değerleri için modelimizi çalıştırıyoruz,
- Her k değeri için datamızdan elde edeceğimiz inertia (ssd) değerleri hesaplanır ve grafik çizdirilir.
- Inertia’da en hızlı düşüşün bittiği dirsek noktasının denk geldiği k değeri optimum k’dır.
- Inertia’nın en minumum olduğu nokta demek gözlem sayısı kadar küme demektir ki bu da zaten kümelemenin mantığına aykırıdır.

**SSD(Sum of Squared Distances)-Inertia**

- Elbow metodunda optimal k’yı belirlemek için kullanılır.
- Her bir data noktasının centroide olan uzaklıklarının karelerinin toplamı alınıp hesaplanır. 
- Inertia’yı minimize etmeye çalışmak hedefimiz. En minimize noktada iterasyon burda durur ve kümeler oluşmuş olur.
- Inertia ne kadar küçükse data noktalarımız centroide o kadar yakındır.

**Silhoutte Coefficient**; kümeleme kalitesini ortaya koyar.

- Labellarımız ve domain knowledge yoksa kullanılır. 

- Önce bir küme içerisindeki bütün gözlemlerin birbirine olan uzaklıklarını bulur. 

- Sonrasında bir küme içerisindeki gözlemlerimizin diğer kümelerimizdeki gözlemlere olan uzaklıklarını bulur.  

- Bu şekilde her bir gözlem için bir skor hesaplanır ve sonuçta her bir küme için ortalama bir skor hesaplanmış olur. 

- 1’e yakınsa iyi bir kümeleme var demektir. 

- Amaç küme içerisindeki mesafenin düşük, kümeler arası mesafenin büyük olmasıdır.

- Inertia küme içi distance ile ilgilenirken Silhoutte kümeler arası distance ile ilgilenir.

```python
# silhouette_score ; yaptığımız kümelemenin kalitesini ölçen bir metriktir, inertia her küme kendi içerisinde centroidi  
# etrafında ne sıklıkta kümelenmiş bunu ölçüyor ( küme içi mesafelere bakıyor yani) Ama bizim ideal clusteringten anlamamız
# gereken bir de maximal inter cluster distance idi  ve silhouette_score işte bu durumu  hesaba katıyor.
# silhouette_score 1'e ne kadar yakınsa kümelemem o kadar kalitelidir.

from sklearn.metrics import silhouette_score
range_n_clusters = range(2,11)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X1)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X1, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

## Why silhouette_score is negative?

- Mavilerin ayrı bir küme, yeşillerin ayrı bir küme olduğunu farzedelim.
- 0 noktasındaki ise yeni gözlemimiz. Bunun hangi kümeye gideceğine karar vereceğiz.
- Sample'ın hangi kümeye atanacağı işlemi yapılırken; her kümenin orta noktası bulunur ve sample hangi kümeye daha yakın ise oraya atanır.
- Mavi noktaların orta noktası 1, yeşillerin orta noktası ise -1'in biraz altı.
- Mavilerin orta noktası sample'a daha yakın olduğu için sample, mavilerin olduğu kümeye atanır.
- Bir çözüm olarak küme sayısı arttırılabilir.

b --> Gözlemin merkeze olan uzaklığı (Burada b=1)

a --> Gözlemin atandığı kümelerdeki elemanlara olan uzaklığının ortalaması (Burada a = √2+√2 = 2√2/2 = √2)

Silhouette Skoru --> (1-√2) / √2 = -0.29

- Yeşillerin olduğu küme kendi içlerinde sıkı bir şekilde kümeleşmişken, mavilerin olduğu küme elemanlarının kendi aralarındaki mesafe oldukça fazla.
- Eğer Silhouette skoru (-) değer alıyorsa, o kümedeki sample'lar birbirleri ile iyi kümeleşememiş, aradaki mesafe uzak demektir. Bu durumda kümeleme sayısının artırılması gerekir.
- Silhouette skorunun 0'a yakın veya 0'dan küçük değer alması demek, küme sayısının artırılması gerektiği demektir.

silhouette_score = (b-a)/max(a,b)

b : the mean nearest-cluster distance 
a : the mean intra-cluster distance 

for red point, 

b = 1 
a = ((1+1)**0.5 + (1+1)**0.5)/2  ==> 1.41

silhouette_score = (1-1.41)/1.41 ==> -0.29

**Hiyerarşik kümeleme**
- Amaç gözlemleri birbirlerine olan benzerliklerine göre kümelerine ayırmaktır.
- Gözlemler daha fazla alt sayıda kümeye ayrılmak istendiğinde kullanılır.
- Hiyerarşik kümeleme yöntemlerinde küme sayısına dendogram sonuçlarına bakarak karar verilirken hiyerarşik olmayan yöntemlerde küme sayısı uygulama yapılmadan önce belirlenir.
- İki metotumuz vardır.

**Agglomerative kümeleme**

- Aşağıdan yukarı - tümevarım. En fazla kullanılan metot.
- Herbir gözleme bakarak benzerliklerine göre birleştire birleştire yukarı çıkar. Sonunda tüm dataya ulaşır.
- Başlangıçta gözlem sayısı kadar küme vardır.
- Veri setinde birbirine en yakın olan iki gözlem bulunur.
- Bu iki nokta bir araya getirilerek yeni bir gözlem oluşturulur. Yani artık veri seti ilk birleşimdeki gözlemlerden oluşmaktadır.
- Aynı işlemler tekrarlanarak yukarı doğru çıkılır. Yani iki kümenin birleşiminden oluşan bu yeni kümeler aynı şekilde birbirlerine benzerliklerine göre tekrar birleştirilir. Bu işlem tüm gözlemler tek bir kümede toplanana kadar tekrar edilir.

**Divisive kümeleme**

- Yukarıdan aşağı - tümdengelim. 
- Yukardan başlayarak tüm datayı sürekli olarak farklılıklarına göre ikiye böle böle kümelemeler birbirinden bağımsız olana kadar iner.
- Başlangıçta tek küme vardır o da veri setidir.
- Tüm gözlemlerin bir arada olduğu küme iki alt kümeye ayrılır.
- Oluşan yeni kümeler birbirlerine benzemeyen alt kümelere bölünür.
- Aynı işlem gözlem sayısı kadar küme elde edilinceye kadar tekrar edilir.

**Not: Karar ağaçlarından farkı; karar ağaçlarında ayırma işlemi hedef değişkene göre yapılırken burada bağımlı değişken olmadığı için gözlemler bağımsız değişkenler üzerinden yapılan uzaklık hesaplarına göre kümelere ayrılır.**

- Hiyerarşik kümelemede başlangıçta küme değerini belirlemek zorunlu değil, ancak KMeans'de zorunlu.
- Ayrıca Hiyerarşik kümelemenin  en büyük dezavantajı predict özelliğinin olmamasıdır. 
- Yeni bir data geldiğinde, datayı veriye ekleyip işlemleri baştan yapmamız gerekiyor.
- Hiyerarşik kümeleme; büyük data setlerinde yavaş olduğundan tercih edilmez.

## Clustering based on Age and Spending Score- X1

```python
from scipy.cluster.hierarchy import linkage

# dendogram benden datayı isterken aynı zamanda mesafelerin neye göre ölçüleceği bilgisini de ister.
# Bu sebeple linkage fonksiyonu içine hem datamızı tanımlayıp hemde ölçümün neye göre yapılacağı belitilerek linkage
# fonksiyonu dendogram içerine verilir. 
# Bu bilgiler ışığında dendogram gözlemleri kümelere atar ve bu atama işlemlerini bir dallanma şeklinde bize sunar.
```

**Hiyerarşik kümelemenin hyperparametrelerine bakarsak**

**affinity** yakınlık hiperparametresi** gözlemler arası mesafenin mesafenin nasıl ölçüleceğini belirleyen parametre. default; euclidean, diğerleri; manhattan, cosine, precomputed. ward kullanılırsa zorunlu olarak euclidean kullanılır.

**linkage** bir gözlemin kümelere nasıl bağlanacağını belirlediğimiz hiperparametre. default; ward, diğerleri; complete, average, single. 

**ward** varyansı düşürecek şekilde gözlemi kümelere bağlar.  Bir gözlem hangi kümenin varyansını daha çok düşürüyorsa o kümeye atanır. 

- Varyans; Varyansın sıfır olması tüm verilerin eşit olduğunu gösterir, yani ortalamadan uzaklaşma yoktur.
- Varyans değerinin artması (büyüklüğü) ilgili verilerin ortalamadan daha fazla uzaklaştığı (dağıldığı) şeklinde yorumlanır. 
- Eğer veri ortalama çevresinde kümelenir ise varyans düşük olur. 
- Eğer veri ortalamadan uzak bir noktaya dağılırsa varyans yüksek olur.

**average** her kümenin ortalaması alınır, yeni gelen gözlem hangi kümenin ortalamasını daha çok düşürüyorsa o kümeye atanır.

**complete** yeni gelen gözlemin her kümenin en uzak gözlemine mesafesi ölçülür ve bu mesafelerin hangisi kısaysa o kümeye atanır.

**single** yeni gelen gözlemin her kümenin en yakın gözlemine mesafesi ölçülür ve bu mesafelerin hangisi kısaysa o kümeye atanır.

```python
hc_ward = linkage(y = X1, method = "ward")    # Varyansa gore atama islemi
hc_complete = linkage(X1, "complete")         # En uzak noktaya gore atama islemi
hc_average = linkage(X1, "average")           # Ortalamaya gore atama islemi
hc_single = linkage(X1, "single")             # En yakin noktaya gore atama islemi

# Bu 4 farkli durum icin, yukarida olusturdugumuz degiskenler ile gorsellestirmemizi yapacagiz
```

```python
from scipy.cluster.hierarchy import dendrogram

# Dendogram, K-means' de kullanilan Elbow metodu gibidir. 
# Dendogram ile bir agaclandirma yapip, kumelerin birbirleri arasindaki mesafeye bakilarak kac kume secilecegine
# karar verilir. 
# Dendogram icine direk datanin kendisini veremiyoruz
```

```python
# burda her bir linkage yöntemi için datamızı kullanarak dendogram çizdiriyoruz. 
# Dendogramı yorumlarken mesafeler dikkate alınır. Kümeler arası mesafe hangisinde fazlaysa onu tercih edeceğiz.
# kümeler içi minimum kümeler arası maksimum mesafeyi ward sağladığından ward ile devam edeceğiz.
# dendogramın mavi çizgi ile gösterdiği bize tavsiye ettiği küme sayısıdır.
#Ancak dendogram kümeler arası mesafeyi max yapacak şekilde bize bir öneride bulunur ama biz yine sadece dendogramdan değil
# scatter plot ve silhouette scoredan da faydalanıp ona göre karar vereceğiz.

# truncate_mode='lastp' ve p=10 ayarlanarak dendogramın sadece son 10 kümeyi bize gösterecek şekilde sadeleştirme yapması
# sağlanır.

plt.figure(figsize = (20,12))

plt.subplot(221)
plt.title("Ward")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_ward, truncate_mode = "lastp", p = 10, show_contracted = True, leaf_font_size = 10)

plt.subplot(222)
plt.title("Complete")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_complete, truncate_mode = "lastp", p = 10, show_contracted = True, leaf_font_size = 10)

plt.subplot(223)
plt.title("Average")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_average, truncate_mode = "lastp", p = 10, show_contracted = True, leaf_font_size = 10)

plt.subplot(224)
plt.title("Single")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_single, truncate_mode = "lastp", p = 10, show_contracted = True, leaf_font_size = 10);

# En fazla uzakligin ward da oldugunu goruyoruz. Bu yuzden ward' i secerek yolumuza devam edecegiz. 
# Ward, 2 kumeleme yapmamiz gerektigini soyluyor. Fakat 4 kumeleme sectigimizde de aradaki mesafelerin hala oldukca uzak
# oldugunu goruyoruz. 
# En uzak mesafe secilecek diye bir kaide yok. Eger guzel bir ayrim saglanabiliyorsa daha fazla kumeleme de yapilabilir.
# Biz kumeleme sayisini 4 olarak belirledik 
```

```python
# linkage değeri olarak ward ı belrlediğimizden daha ayrıntılı inceleyebilmek için tekrar çizdiriyoruz.

plt.figure(figsize = (20,8))
plt.title("Dendrogram")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_ward, truncate_mode = "lastp", p = 10, show_contracted = True, leaf_font_size = 10);
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
# burda ise silhouette skorlarımızı hesaplayıp dendogramdan elde ettiğimiz insightlarla karşılaştırıp
# küme sayımıza karar vereceğiz

from sklearn.metrics import silhouette_score
range_n_clusters = range(2,11)
for num_clusters in range_n_clusters:
    # intialise kmeans
    Agg_model = AgglomerativeClustering(n_clusters=num_clusters)
    Agg_model.fit(X1)
    cluster_labels = Agg_model.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X1, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
    
# burda en yüksek skor 4 olduğundan cluster sayısı olarak dörtle devam edeceğiz
```
